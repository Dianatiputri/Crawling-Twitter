```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import nltk 
import string
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
#Cleaning Text
def remove_punct(text):
    #text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub(r'[^a-zA-Z0-9]', ' ', str(text))
    text = re.sub(r'\b\w{1,2}\b', '', text)  # menghilangkan 2 kata
    text = re.sub(r'\s\s+', ' ', text)
    return text

df['ISI'] = df['isi'].apply(lambda x: remove_punct(x))
```

```python
def cleaning_text(text):
    text = re.sub(r'[@]',r' ',text)
    return text
df['author_name'] = df['author'].apply(lambda x: cleaning_text(x))
df.head(5)
```

```python
#Tokenization

def tokenization(text):
    text = re.split('\W+', text)
    return text

df['TOKENIZATION'] = df['ISI'].apply(lambda x: tokenization(x.lower()))
df.head(5)
```

```python
#Stop Removal

stopword = nltk.corpus.stopwords.words('indonesian')


def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
df['STOP_REMOVAL'] = df['TOKENIZATION'].apply(lambda x: remove_stopwords(x))
df.head(5)
```

```python
#Case Folding
df['isi'] = df['ISI'].str.lower()
df['user_name'] = df['user'].str.lower()
df['author'] = df['author_name'].str.lower()
df.head(5)
```

```python
stop_removal = df[['STOP_REMOVAL']]

def fit_stopwords(text):
    text = np.array(text)
    text = ' '.join(text)
#     text = text.tostring()
#     text = str(text)
    return text
    
df['STOP_REMOVAL'] = df['STOP_REMOVAL'].apply(lambda x: fit_stopwords(x))
df.head(5)
```
